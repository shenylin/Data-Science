```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd
import seaborn as sns 
from sklearn.metrics import silhouette_score
```

```python
df = pd.read_csv(r'Auto.csv')             
print (df)
```

```python
df
```

```python
type(df)
```

```python
df['Weight'].dtypes
```

```python
auto = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(df)
    auto.append(kmeans.inertia_)
```

```python
sns.set()
plt.plot(range(1, 11), auto)
plt.title('Selecting the Number of Clusters using the Elbow Method')
plt.xlabel('Clusters')
plt.ylabel('Within Cluster Sum of Squares')
plt.show()
```

```python
kmeans = KMeans(n_clusters=2).fit(df)
print("Cluster Centroid Values:")
centroids = kmeans.cluster_centers_
print(centroids)
label = kmeans.fit_predict(df)
#print(label, df)
kmeans_silhouette = silhouette_score(df, label)
print("Overall Average Silhouette Score:")
print(kmeans_silhouette)
```

```python
kmeans = KMeans(n_clusters=3).fit(df)
print("Cluster Centroid Values:")
centroids = kmeans.cluster_centers_
print(centroids)
label = kmeans.fit_predict(df)
#print(label, df)
kmeans_silhouette = silhouette_score(df, label)
print("Overall Average Silhouette Score:")
print(kmeans_silhouette)
```

```python
kmeans = KMeans(n_clusters=4).fit(df)
print("Cluster Centroid Values:")
centroids = kmeans.cluster_centers_
print(centroids)
label = kmeans.fit_predict(df)
#print(label, df)
kmeans_silhouette = silhouette_score(df, label)
print("Overall Average Silhouette Score:")
print(kmeans_silhouette)
```

```python
kmeans = KMeans(n_clusters=5).fit(df)
print("Cluster Centroid Values:")
centroids = kmeans.cluster_centers_
print(centroids)
label = kmeans.fit_predict(df)
#print(label, df)
kmeans_silhouette = silhouette_score(df, label)
print("Overall Average Silhouette Score:")
print(kmeans_silhouette)
```
