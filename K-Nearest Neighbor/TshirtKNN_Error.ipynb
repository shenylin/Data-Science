```python
# -*- coding: utf-8 -*-
"""
Created on Tue Feb 22 11:32:48 2022
@author: Renee
"""
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
```

```python
#load dataset
Tshirt = pd.read_csv('/Users/chenghsuanlin/Downloads/Tshirt.csv')
#Tshirt = pd.read_csv(r'C:\Users\Renee\Desktop\Tshirt.csv')
print(Tshirt)
```

```python
#Scatter Plot
plt.scatter(Tshirt['Height'], Tshirt['Weight'])
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()
#Set Features and Target Variable
X = Tshirt[['Height', 'Weight']]
Y = Tshirt['T Shirt Size']
```

```python
#Determine classifier error rate
error_rate = []
for i in range(1,10):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X, Y)
    #Y_pred=knn.predict(X)
    pred_i = knn.predict(X)
    error_rate.append(np.mean(pred_i != Y))
    
plt.figure(figsize=(10,6))
plt.plot(range(1,10),error_rate,color='blue', linestyle='dashed', marker='o',
markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```
