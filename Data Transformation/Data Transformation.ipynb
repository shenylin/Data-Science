### Normalize a Dataset Without Converting Columns to Array

```python
admission = pd.read_csv("/Users/chenghsuanlin/downloads/Admission.csv")
admission_1 = preprocessing.normalize(admission)
scaled_df = pd.DataFrame(admission_1)
scaled_df.head()
```

```python
# Normalize the dataset using MinMaxScaler
scaler = preprocessing.MinMaxScaler()
names = admission.columns
dataframe = scaler.fit_transform(admission)
scaled_df = pd.DataFrame(dataframe, columns=names)
scaled_df.head()
```
